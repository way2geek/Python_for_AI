Write a Python program to generate N times of die rolls where N should be updated and adjusted to get the desired result.  
Use Matplotlib to visualize the result and proof that the probability of each outcome (e.g. get the number '1') is approximately *0.16* - *0.17*

```python
# Try with different number of N

import random
n1 = 100
n2 = 1000
n3 = 100000000
nums = {}
for i in range(1,7):
    nums[i]=0

# Please write your code here
for i in range(n3):
    x= random.randint(1, 6)
    nums[x]+=1

for i in range(1,7):
    nums[i] = nums[i]/n3

x = list(nums.keys())
y= list(nums.values())
plt.ylim((0.15,0.17))
plt.bar(x,y , color="blue")
plt.show()
```

The easiest way to make a live animation in Matplotlib is to use one of the Animation classes.  
[matplotlib.animation](https://matplotlib.org/stable/api/animation_api.html)  

Please plot an animation for 

$y = sin(x)$

```python
import math as  m 
import matplotlib.pyplot as plt
from IPython.display import clear_output
from matplotlib import animation
import numpy as np
fig = plt.figure()
x= np.arange(0,10*m.pi,m.pi/4)
y =np.sin(x)
# Please write your code here
ax = fig.subplots()
l = ax.plot(x,y)
l = l[0]
def animate(i):
    l.set_data(x[:i],y[:i])
    return 1
for i in range(len(x)):
    animate(i)
    clear_output(wait=True)
    display(fig)
```

Plot this figure with the given theta values:

$x = cos(\theta)$

$y = sin(\theta) + cos(\theta)^{2/3}$

You can define $\theta$ using the way we have learned for hart shape in our lecture notes.


```python
# Please write your code here
import numpy as np

from matplotlib import pyplot as plt
theta = np.linspace(0,2*np.pi,100)
x= np.cos(theta)
m = np.cos(theta)**2
m = np.power(m,1/3)
y =np.sin(theta)+m
plt.plot(x,y)
plt.title('Heart')
plt.show()
```
