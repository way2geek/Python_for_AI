Write a class for Person

Basic Properties: Age, Name, Sex. 

Extension Properties: Working, Sleepings, just consider the hours they work and sleep everyday. 

Then, instantiate the two classes into to different persons


```python
class Person:
    def __init__(self, name, age, sex):
        self.name = name
        self.age = age
        self.sex = sex

    def Working(self, hours):
        print(self.name + 'working' + str(hours) + "hours every day")

    def Sleeping(self, hours):
        print(self.name + 'Sleeping' + str(hours) + "hours every day")
    
p1 = Person("John", 36,'male')
p2 = Person('Eric',23,'male')
p1.Working(8)
p1.Sleeping(9)
p2.Working(9)
p2.Sleeping(7)
print(p1.name)
print(p1.age)
print(p2.name)
print(p2.age)
```

### Statistical estimation in seaborn goes beyond descriptive statistics. 
### For example, it is possible to enhance a scatterplot by including a linear regression model (and its uncertainty) using lmplot():

```python
exercise['time'].replace(['1 min', '15 min','30 min'],
                        [1,15,30], inplace=True)
sns.lmplot(data=exercise, x="time", y="pulse", col="kind", hue="diet");
```

```python
exercise.loc[tips['pulse'] > 120]
```

```python
@interact
def show_pulse(column='pulse', x=120):
    return exercise.loc[tips[column] > x]
```

```python
@interact
def show_pulse(column=['pulse','kind','time','diet'], x=120):
    return exercise.loc[tips[column] > x]
```

```python
@interact
def correlation(column1 = list(exercise.select_dtypes('number').columns),
                column2 = list(exercise.select_dtypes('number').columns)):
    print(f"Correlation: {exercise[column1].corr(exercise[column2])}")
```
