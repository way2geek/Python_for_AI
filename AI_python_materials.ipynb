### Defining function

Using def, () and return 

Variables passed into function

Values returned by function



```python
import math

def move(x, y, step, angle=0):
    nx = x + step*math.cos(angle)
    ny = y - step*math.sin(angle)
    return nx, ny
```

```python
new_x, new_y = move(5, 10, 1.0, angle=10)
print(new_x, new_y)
```

The computer will repeatly do operations in the code blocks,

#### while loop

The computer will repeatly do operations in the code blocks,
Until the condition is not satisfied.



```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
```

```python
for x in "banana":
  print(x)
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x) 
  if x == "banana":
    break
```

### Class
We can define classes ourselves.

```python
class Foo():
    class_variable = 'same' # class variable

    def __init__(self, x):
        # first parameter "self" for instance reference
        self.x = x
    
    # instance method
    def print_x(self):
        print(self.x)
        
    # class method
    @classmethod
    def modify_class_variable(cls):
        cls.class_variable = 'changed'

    # static method
    @staticmethod
    def print_hello():
        print("Hello!")

print(Foo.class_variable)    

obj1 = Foo(6)
obj1.print_x()
print(obj1.class_variable)

obj2 = Foo(5)
obj2.print_x()
print(obj2.class_variable)

obj1.modify_class_variable()
print(obj1.class_variable)
print(obj2.class_variable)
print(Foo.class_variable)
```

inherit classes:

```python
# Inherits variables and methods
class Bar(Foo):
    pass

obj = Bar(3)
obj.print_x()
```
