```python
class Dog:
  def __init__(self, breed, age):
    self.breed = breed
    self.age = age
  def Eating(self, times):
    print(self.breed + ' Eating ' + str(times) + " times every day") 
  def Sleeping(self, hours):
    print(self.name + ' Sleeping ' + str(hours) + " hours erery day")

object1 = Dog("Bulldog", 4)
object2 = Dog("Chihuahua", 2)
object3 = Dog("Husky", 6)

print(object1)
print(object2)

object1.Eating(5)
object2.Eating(4)
object3.Eating(3)
```
